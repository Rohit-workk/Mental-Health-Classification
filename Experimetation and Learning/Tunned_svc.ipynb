```python
# lets import the requirements
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# lets import the data
df = pd.read_csv("processed_data.csv")
df.head()
```

```python
# lets drop the columns
df.drop(columns = "Unnamed: 0",inplace = True)
```

```python
# here we work with support vector machine
```

```python
# lets make a split of data 
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df.drop(columns="status"),df["status"],test_size=0.2,stratify=df["status"],random_state=1)
```

```python
print(xtrain.shape)
```

```python
# here we use the tfidf for text encoding as this outperforms the bag of words
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

from sklearn.feature_extraction.text import TfidfVectorizer

num_cols = ["statement_length", "num_words", "vocabulary_size", "avg_word_length"]

preprocessor = ColumnTransformer([
    ("text", TfidfVectorizer(max_features=15000), "clean_text"),   # preprocessed text
    ("num", StandardScaler(), num_cols)             # numeric features
])
```

```python
# lets make a pipeline
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC
pipeline_svc = Pipeline([
    ("preprocess", preprocessor),
    ("model", LinearSVC(class_weight="balanced",max_iter=5000,dual="auto"))
])

```

```python
# lets make the parameter grid

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

param_dist = {
    "model__C": uniform(0.1, 10)
}

```

```python

# RandomizedSearchCV
random_search_svc = RandomizedSearchCV(
    pipeline_svc,
    param_distributions=param_dist,
    n_iter=20,
    cv=3,
    scoring="f1_macro",   # important
    n_jobs=-1,
    verbose=2,
    random_state=42
)

```

```python
# lets fit it on our data
random_search_svc.fit(xtrain, ytrain)

```

```python
# lets check the best parameters
best_svc = random_search_svc.best_estimator_

print("Best Params:", random_search_svc.best_params_)
print("Best Score:", random_search_svc.best_score_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = best_svc.predict(xtest)

print(classification_report(ytest, y_pred))

cm = confusion_matrix(ytest, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Purples")
plt.title("LinearSVC - Confusion Matrix")
plt.show()
```

```python
# lets now also tune the vectorizer also 
```

```python
# lets again modify the pipeline

feature_builder = ColumnTransformer([
    ("text_features", TfidfVectorizer(), "clean_text"),
    ("numeric_features", StandardScaler(), num_cols)
])

```

```python
svm_pipeline = Pipeline([
    ("feature_builder", feature_builder),
    ("classifier", LinearSVC(
        class_weight="balanced",
        max_iter=5000,
        dual="auto"
    ))
])
```

```python
# Lets make a parameters grid
from scipy.stats import uniform, randint

param_dist = {
    # TF-IDF tuning
    "feature_builder__text_features__max_features": randint(5000, 20000),
    "feature_builder__text_features__ngram_range": [(1,1), (1,2)],
    "feature_builder__text_features__min_df": randint(2, 6),
    "feature_builder__text_features__max_df": [0.8, 0.9, 1.0],
    
    # SVM tuning
    "classifier__C": uniform(0.1, 10)
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

svm_search = RandomizedSearchCV(
    svm_pipeline,
    param_distributions=param_dist,
    n_iter=30,
    cv=3,
    scoring="f1_macro",
    n_jobs=-1,
    verbose=2,
    random_state=42
)
```

```python
svm_search.fit(xtrain, ytrain)
```

```python
# lets check the best parameters
best_svm = svm_search.best_estimator_
print("Best Params:", svm_search.best_params_)
print("Best Score:", svm_search.best_score_)
```

```python

# evalution of model
y_pred = best_svm.predict(xtest)

print(classification_report(ytest, y_pred))

cm = confusion_matrix(ytest, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Purples")
plt.title("LinearSVC - Confusion Matrix")
plt.show()
```
